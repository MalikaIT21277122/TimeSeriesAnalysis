```python
import pandas as pd
import os

# Function to load datasets
def load_data():
    # Replace the paths with your actual dataset paths
    defect_facts = pd.read_csv("CESAW_defect_facts.csv")
    component_data = pd.read_csv("CESAW_Component_Data.csv")
    time_facts = pd.read_csv("CESAW_time_fact.csv")
    task_facts = pd.read_csv("CESAW_task_fact.csv")
    size_facts = pd.read_csv("CESAW_size_facts.csv")
    project_summary = pd.read_csv("CESAW_project_summary.csv")
    project_data = pd.read_csv("CESAW_Project_Data.csv")

    return defect_facts, component_data, time_facts, task_facts, size_facts, project_summary, project_data

# Function to clean datasets
def clean_data(defect_facts, component_data, time_facts, task_facts, size_facts, project_summary, project_data):
    # Print column names of defect_facts to inspect
    print(f"Defect Facts Columns: {defect_facts.columns}")

    # Handling date columns with custom format (MM/DD/YYYY)
    project_summary['Start Date'] = pd.to_datetime(project_summary['Start Date'], format='%m/%d/%Y', errors='coerce')
    project_summary['End Date'] = pd.to_datetime(project_summary['End Date'], format='%m/%d/%Y', errors='coerce')

    # Drop rows with missing key columns in project_summary
    project_summary.dropna(subset=['Start Date', 'End Date'], inplace=True)

    # Check if 'defect_type_key' column exists before attempting to clean it
    if 'defect_type_key' in defect_facts.columns:
        defect_facts.dropna(subset=['defect_type_key'], inplace=True)
        defect_facts['defect_type_key'] = defect_facts['defect_type_key'].astype('category')

    # Check if 'size' exists in component_data and handle it
    if 'size' in component_data.columns:
        component_data['size'] = pd.to_numeric(component_data['size'], errors='coerce')

    # Returning cleaned datasets
    return defect_facts, component_data, time_facts, task_facts, size_facts, project_summary, project_data

# Function to save cleaned data
def save_cleaned_data(defect_facts, component_data, time_facts, task_facts, size_facts, project_summary, project_data):
    # Ensure directory exists
    output_dir = 'cleaned_data'
    os.makedirs(output_dir, exist_ok=True)

    # Save cleaned data to CSV
    defect_facts.to_csv(os.path.join(output_dir, 'cleaned_defect_facts.csv'), index=False)
    component_data.to_csv(os.path.join(output_dir, 'cleaned_component_data.csv'), index=False)
    time_facts.to_csv(os.path.join(output_dir, 'cleaned_time_facts.csv'), index=False)
    task_facts.to_csv(os.path.join(output_dir, 'cleaned_task_facts.csv'), index=False)
    size_facts.to_csv(os.path.join(output_dir, 'cleaned_size_facts.csv'), index=False)
    project_summary.to_csv(os.path.join(output_dir, 'cleaned_project_summary.csv'), index=False)
    project_data.to_csv(os.path.join(output_dir, 'cleaned_project_data.csv'), index=False)

# Main function to run the cleaning process
def main():
    defect_facts, component_data, time_facts, task_facts, size_facts, project_summary, project_data = load_data()
    defect_facts, component_data, time_facts, task_facts, size_facts, project_summary, project_data = clean_data(
        defect_facts, component_data, time_facts, task_facts, size_facts, project_summary, project_data
    )
    save_cleaned_data(defect_facts, component_data, time_facts, task_facts, size_facts, project_summary, project_data)

# Run the main function
if __name__ == "__main__":
    main()
```
